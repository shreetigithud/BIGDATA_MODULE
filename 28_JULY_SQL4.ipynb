```python
1.  Write a Python program that uses the HiveQL language to create a table named "Employees" with columns for "id," "name," and "salary."
```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Create the Employees table using HiveQL
create_table_query = """
    CREATE TABLE IF NOT EXISTS Employees (
        id INT,
        name STRING,
        salary DECIMAL(10, 2)
    )
"""
cursor.execute(create_table_query)

# Show the table schema
show_table_query = "DESCRIBE Employees"
cursor.execute(show_table_query)
table_schema = cursor.fetchall()

print("Employees Table Schema:")
for column in table_schema:
    print(column)

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
2.  Create a Python program that retrieves records from a Hive table named "Customers" where the age is greater than 30.
```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Execute the query to retrieve records
query = "SELECT * FROM Customers WHERE age > 30"
cursor.execute(query)

# Fetch and process the results
results = cursor.fetchall()
for row in results:
    # Access the columns using index or column name
    customer_id = row[0]
    name = row[1]
    age = row[2]
    email = row[3]

    # Print the retrieved data
    print(f"Customer ID: {customer_id}, Name: {name}, Age: {age}, Email: {email}")

# Close the cursor and connection
cursor.close()
conn.close()

```

```python
3.  Write a Python script that sorts records in descending order based on the "timestamp" column in a Hive table named "Logs."
```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Execute the query to sort records in descending order
query = """
    SELECT * 
    FROM Logs 
    ORDER BY timestamp DESC
"""
cursor.execute(query)

# Fetch and process the results
results = cursor.fetchall()
for row in results:
    # Access the columns using index or column name
    timestamp = row[0]
    log_message = row[1]

    # Print the retrieved data
    print(f"Timestamp: {timestamp}, Log Message: {log_message}")

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
5.  Write a Python script that calculates the average salary of employees from a Hive table named "Employees".
```

```python
6.  Implement a Python program that uses Hive partitioning to create a partitioned table named "Sales_Data" based on the "year" and "month" columns.
```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Execute the query to create the partitioned table
query = """
    CREATE TABLE Sales_Data (
        id INT,
        sales_amount FLOAT
    )
    PARTITIONED BY (year INT, month INT)
"""
cursor.execute(query)

# Execute the query to add partitions to the table
add_partitions_query = """
    ALTER TABLE Sales_Data
    ADD PARTITION (year=2022, month=1)
    ADD PARTITION (year=2022, month=2)
    ADD PARTITION (year=2022, month=3)
"""
cursor.execute(add_partitions_query)

# Show the table description with partitions
show_table_query = "DESCRIBE FORMATTED Sales_Data"
cursor.execute(show_table_query)
table_description = cursor.fetchall()

print("Sales_Data Table Description:")
for row in table_description:
    print(row)

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
8.  Create a Python program that performs an inner join between two Hive tables, "Orders" and "Customers," based on a common column.
```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Execute the inner join query
query = """
    SELECT o.*, c.customer_name
    FROM Orders o
    JOIN Customers c ON o.customer_id = c.customer_id
"""
cursor.execute(query)

# Fetch and process the results
results = cursor.fetchall()
for row in results:
    # Access the columns using index or column name
    order_id = row[0]
    customer_id = row[1]
    order_date = row[2]
    customer_name = row[3]

    # Print the retrieved data
    print(f"Order ID: {order_id}, Customer ID: {customer_id}, Order Date: {order_date}, Customer Name: {customer_name}")

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
9.  Implement a Python program that uses the Hive SerDe library to process JSON data stored in a Hive table named "User_Activity_Logs."

```

```python
from pyhive import hive

# Create a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your-username')

# Create a cursor
cursor = conn.cursor()

# Set the SerDe properties for JSON processing
serde_properties = {
    'serialization.format': '1',
    'json.serialize.decimal.precision': '10',
    'json.serialize.decimal.divisor': '100'
}

# Enable the JSON SerDe
cursor.execute("SET hive.default.jsonserde=")

# Create a Hive table using the JSON SerDe
create_table_query = """
    CREATE TABLE IF NOT EXISTS User_Activity_Logs (
        user_id INT,
        activity_name STRING,
        activity_time TIMESTAMP
    )
    ROW FORMAT SERDE 'org.apache.hive.hcatalog.data.JsonSerDe'
    WITH SERDEPROPERTIES (
        'serialization.format' = '1',
        'json.serialize.decimal.precision' = '10',
        'json.serialize.decimal.divisor' = '100'
    )
    STORED AS TEXTFILE
"""
cursor.execute(create_table_query)

# Query the Hive table
query = "SELECT * FROM User_Activity_Logs"
cursor.execute(query)

# Fetch and process the results
results = cursor.fetchall()
for row in results:
    user_id = row[0]
    activity_name = row[1]
    activity_time = row[2]
    print(f"User ID: {user_id}, Activity Name: {activity_name}, Activity Time: {activity_time}")

# Close the cursor and connection
cursor.close()
conn.close()
```
